# P-Curve: A Key to the File-Drawer

## Uri Simonsohn et al., 2013, American Psychological Association

> Because a repeatedly obtained false-positive finding
would require implausibly large file-drawers of failed attempts, it
would seem that one could be confident that findings with multi-
study support are in fact true



__ this is the definition of p-hacking:  __


> While collecting and analyzing data, researchers have many 
decisions to make, including whether to collect more data, 
which outliers to exclude, which measure(s) to analyze, which 
covariates to use, and so on. If these decisions are not made 
in advance but rather are made as the data are being analyzed, 
then researchers may make them in ways that self-servingly 
increase their odds of publishing (Kunda, 1990)

__ this is the consequence of p-hacking:  __

> p-hacking can allow researchers to get most studies to reveal 
significant relationships between truly unrelated variables

> all p values below .05 are potentially publishable and
hence observable. This allows us to make unbiased inferences
from the distribution of p values that fall below .05, i.e., from
p-curve.


__The distribution of p values under the null is uniform__

Let F(X) be the cumulative density function. 

$$ p_{\alpha} = 1 - F(t_{\alpha}) $$

Let's look at how is F(X) distributed :

$$ F(t) = P(X \leq t) = P(F(X) \leq F(t)) $$

So, now we have 
$$ P(F(X) \leq u) = u $$

Which is the definition of a uniform random variable. 

> When a studied effect does exist (i.e., the null is false), the
expected distribution of p values of independent tests is right-
skewed, that is, more ps < .025 than .025 < ps < .05


> given a type I error risk, p-curve’s shape is a function only of effect size and sample size


> 
* Sets of studies investigating effects that exist are expected to
produce right-skewed p-curves.
* Sets of studies investigating effects that do not exist are
expected to produce uniform p-curves.
* Sets of studies that are intensely p-hacked are expected to
produce left-skewed p-curves.

## Method

### Does a Set of Studies Contain Evidential Value? Testing for Right-Skew 

> First, one computes, for each significant p value, the 
probability of observing a significant p value _at least as extreme_ if 
the null were true. 

ie. $P(p_u \leq p_{obs})$ 

This is the pp-value.
If $p_obs = .01$ there is 20% chance that a significant p value drawn from a uniform distribution will be less than  .01; hence pp-value = .2 (remember : we have values between 0 and 0.05)

$P(p_u \leq 0.01) = .2$

The **second step** is to aggregate the pp values, which we do using Fisher’s method. This yields an overall $\chi2$ test for skew,
   with twice as many degrees of freedom as there are p values.

Fisher technique: 

* The logarithm of uniform is exponential
* An exponential with rate 1/2 is chi-squared
* The sum of chi-squared is also chi-squared

> For example, imagine a set of three studies, with p values of .001, .002, and .04, and thus pp values of .02, .04, and .8. Fisher’s method generates an aggregate, $\chi$2(6) = 14.71, p = .022, a significantly right-skewed p-curve


### Does a Set of Studies Lacks Evidential Value? Testing for Power of 33% 

when a p-curve is not significantly right-skewed: 

* the effect we are trying to detect is very small
* the measurements are too noisy

__Not testing that the effect observed in studies is zero, but that it is very small__



> When p-curve is not significantly right-skewed, we propose testing whether it is flatter than one would expect if studies were powered at 33%


$$ P(p_{33pwr} > P_{obs}) $$

> For example, we know that t tests with 38 degrees of freedom and powered at 33% will result in 57.6% of significant p values being > .01, and so pp value corresponding to p = .01 for a t(38) test would be pp = .576. 


> Only 10.4% of significant p values are expected to be greater than .04  so the pp value corresponding to p = .04 for this test would be pp = 0.104


> Under the null of 33% power, observed values of .040, .045, and .049 lead to pp values of
.104, .0502, and .0098, respectively, leading to an overall $\chi$2(6) = 19.76, p = .003;
